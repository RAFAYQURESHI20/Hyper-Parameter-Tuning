# **Load Dataset**

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head())
```

```python
df['flower']=iris.target
df['flower']=df['flower'].apply(lambda x: iris.target_names[x])
df
```

# **Cross Validation**

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
cross_val_score(SVC(kernel='linear',C=10,gamma='auto'), iris.data, iris.target, cv=3)
```

```python
cross_val_score(SVC(kernel='rbf',C=10,gamma='auto'), iris.data, iris.target, cv=3)
```

# **Find the Average with scratch**

```python
import numpy as np
kernal=['linear','rbf']
c=[1,10,20]
avg_score={}
for k_val in kernal:
 for c_val in c:
  c_score=cross_val_score(SVC(kernel=k_val,C=c_val,gamma='auto'), iris.data, iris.target, cv=3)
  avg_score=[k_val+' '+str(c_val),np.average(c_score)]
  print(avg_score)
```

# **find the average with gridsearchcv**

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=3,return_train_score=False)
```

```python
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

# **Hyper Parameter Tuning**

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
```

```python
models_params={'svm':{
    'model':svm.SVC(gamma='auto'),
    'params':{
        'C':[1,10,20],
        'kernel':['rbf','linear']
    }
},
               'random_forest':{
                   'model':RandomForestClassifier(),
                   'params':{
                       'n_estimators':[1,5,10]
                   }
               },
               'logistic_regression':{
                   'model':LogisticRegression(solver='liblinear',multi_class='auto'),
                   'params':{
                       'C':[1,5,10]
                   }
               }

}

```

```python
scores=[]
for model_name,mp in models_params.items():
  clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
  clf.fit(iris.data,iris.target)
  scores.append({
      'model':model_name,
      'best_score':clf.best_score_,
      'best_params':clf.best_params_
  })
```

```python
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
